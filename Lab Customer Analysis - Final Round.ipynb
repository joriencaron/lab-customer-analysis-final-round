## Cleaning/Wrangling - Round 2

```python
import pandas as pd
import numpy as np

#Show the dataframe shape.
df_marketing = pd.read_csv('marketing_customer_analysis.csv')
df_marketing.head()
```

```python
#Standardize columns, deal with NaN values, change date, create new columnf for month

def clean_data(df, column, new_column):
    #df = df.drop([column], axis=1)
    cols = []
    for c in df.columns:
        cols.append(c.lower())
    df.columns = cols
    perc = 25.0
    min_count =  int(((100-perc)/100)*df.shape[0] + 1)
    df = df.dropna( axis=1, 
                thresh=min_count)
    numerical = df.select_dtypes(include= np.number)
    for c in numerical:
        if numerical[c].isna().sum() > 0:
            numerical[c] = numerical[c].fillna(np.mean(numerical[c]))
    categorical = df.select_dtypes(include= np.object)
    for c in categorical.columns:
        df[c].fillna(df[c].mode()[0], inplace=True)
    df[column] = pd.to_datetime(df[column], errors='coerce')
    df[new_column] = df[column].dt.month
    return df

clean_data = clean_data(df_marketing,'effective to date','month').head()

clean_data.head()
```

## Reporting - Conclusions based on analysis

After thoroughly analysing the given dataset, several conclusions and recommendations can be drawn.
Mainly, 3 variables influence the total claim amount of the customers:

1. Customer lifetimevalue:
This variable correlates positively with the total claim amount, which means the higher the lifetime value the higher the total claim amount. This correlation however is small (.23), implying that if the customer's lifetime value goes up their claim amount only goes up by a little. This should not be considered as a problem, as their value as a customer is increasing which weighs out the extra cost for the company in claim amount form that partical customer.

2. Income
Income actually negatively correlates with the total claim amount, implying that richer customers claim less. This is positive for the company, because these customers will make them more profitable, as they mainly pay for the insurance but claim less. 

3. Monthly premium auto
This variable also correlates positively with total claim amount, implying that the higher the premium the higher the claim amount. This makes sense as customers with higher premiums probably take this coverage as they experience more problems with their cars. Customers paying a higher a premium are therefore less interesting in profit making, if their premium doesn't make up for their claims. Interestingly, Monhtly premium auto and customer lifetime value also positvely correlate with each other implying that the higher the premium the higher the customers worth. However, its unclear in this dataset if the total claim amount gets substracted from the customer lifetime value, making it difficult to make clearcut conclusions about this. 

Recommendations:

Based on this analysis the company most profitable customers are people with a high income, that have a lower premium and lower customer lifetime value. 
Assuming the total claim amount gets subtracted from the customer lifetime value, a higher monthly premium and lifetime value might not be a bad thing, because the value grows accordingly with the higher claim amounts. - due to a lack of info this cannot be proven unfortunately. However, for this car insurance the big wins are to be found in customers with a high income or a lower premium. It's advisable for this company to spend marketing budget on attracting new high income customers in combination with lower premiums. It would be intersting as well for the company to convert current customers in paying lower premiums.
